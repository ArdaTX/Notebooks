# Importación de Librerias

```python
import numpy as np # For algebra lineal
import pandas as pd # Para leer datos (archivo csv) y crear un marco de datos
import matplotlib.pyplot as plt # Para trazar un gráfico
import seaborn as sns # Para visualizar datos
import plotly.express as px # Para visualizar gráficos en 3D
```

# Lectura del archivo

```python
df = pd.read_csv('player_stats.csv', encoding='ISO-8859-1')
```

**Descripción de columnas**

- Player: El nombre del futbolista.
- Country: La nacionalidad o país de origen del jugador.
- Height: La altura del jugador en centímetros.
- Weight: El peso del jugador en kilogramos.
- Age: La edad del jugador.
- Club: El club al que está afiliado actualmente el jugador.
- Ball Control: La habilidad del jugador para controlar el balón.
- Dribbling: La habilidad del jugador en el regate.
- Marking: Habilidad de marcaje del jugador.
- Slide Tackle: Habilidad del jugador para realizar entradas deslizantes.
- Stand Tackle: Habilidad del jugador para realizar placajes de pie.
- Aggression: Nivel de agresividad del jugador.
- Reactions: Tiempo de reacción del jugador.
- Attacking Position: Posicionamiento del jugador en las jugadas de ataque.
- Interceptions: Habilidad del jugador para interceptar pases.
- Vision: Visión del jugador en el campo.
- Composure: Compostura del jugador bajo presión.
- Crossing: Capacidad del jugador para realizar centros.
- Short Pass: Precisión del jugador en los pases cortos.
- Long Pass: Habilidad del jugador en el pase largo.
- Acceleration: Aceleración del jugador en el campo.
- Stamina: Nivel de resistencia del jugador.
- Strength: Fuerza física del jugador.
- Balance: Equilibrio del jugador durante el juego.
- Sprint Speed: Velocidad del jugador en los sprints.
- Agility: Agilidad del jugador en las maniobras.
- Jumping: Capacidad de salto del jugador.
- Heading: Habilidad de cabeceo del jugador.
- Shot Power: Potencia de disparo del jugador.
- Finishing: Habilidad de remate del jugador.
- Long Shots: Capacidad del jugador para realizar tiros de larga distancia.
- Curve: Habilidad del jugador para curvar el balón.
- Free Kick Accuracy: Precisión del jugador en situaciones de tiro libre.
- Penalties: Habilidad del jugador para lanzar penaltis.
- Volleys: Habilidad de volea del jugador.
- Goalkeeper Positioning: Atributo de posicionamiento del portero (específico para porteros).
- Goalkeeper Diving: Habilidad de clavada del portero (específico para porteros).
- Goalkeeper Handling: Habilidad de manejo de balón del portero (específico para porteros).
- Goalkeeper Kicking: Habilidad del portero para chutar (específico para porteros).
- Goalkeeper Reflexes: Reflejos del portero (específico para porteros).
- Value: El valor estimado del jugador.

```python
# Función para eliminar '$' y '.' de una cadena
def f(x):
    return float(x.replace('$','').replace('.',''))
```

```python
f('$1.400.00')
```

```python
# Como el valor es numérico pero debido a $ y "," pandas lo está tratando como objeto
# vamos a convertirlo a float
df['value'] = np.vectorize(f)(df['value'])
```

```python
# diagrama de caja
sns.boxplot(data=df,x='weight')
```

**Analisis Bivariante**

```python
sns.pairplot(data=df[['gk_positioning', 'gk_diving', 'gk_handling', 'gk_kicking',  #grafico de pares
       'gk_reflexes']], diag_kind='kde')
# podemos ver que estas 5 columnas parecen estar bastante relacionadas entre sí
```

```python
sns.jointplot(data=df,x='height',y='weight',kind='hex') # dispersión
```

```python
# los jugadores con más agresividad parecen tener más habilidad para cabecear
sns.scatterplot(data=df,x='aggression',y='heading')
```

**Analisis Multivariable**

```python
# visualicemos 'slide_tackle', 'stand_tackle' e 'interceptions'
# estos tres parecen estar correlacionados entre sí
# el color representa el valor del jugador
fig = px.scatter_3d(df, x='slide_tackle', y='stand_tackle', z='interceptions',
              color='value')
fig.show()
```

```python
# aquí visualizando 'pase_largo', 'pase_corto' y 'resistencia' de los jugadores
# podemos ver que los jugadores con valores altos de estas variables parecen ganar más
fig = px.scatter_3d(df, x='long_pass', y='short_pass', z='stamina',
              color='value')
fig.show()
```

```python
fig = px.scatter_3d(df, x='crossing', y='ball_control', z='dribbling',
              color='value')
fig.show()
```

```python
# visualicemos peso, altura y edad en 3d plot
fig = px.scatter_3d(df, x='weight', y='height', z='age',
              color='value')
fig.show()
```
